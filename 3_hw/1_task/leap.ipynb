```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
```

```python
rho = [1.1, 1.2]

df = defaultdict(pd.DataFrame)
for r in rho:
    df[r] = pd.read_csv(f"1_thermo_{r}.txt", header=25, nrows=1501, sep='\s+')

df[1.2]
```

```python
colors = ["tab:blue", "crimson"]

for i, r in enumerate(rho):
    temp = df[r].filter(['Temp']).to_numpy().astype(float)
    toteng = df[r].filter(['TotEng']).to_numpy().astype(float)
    plt.scatter(temp, toteng, s=0.1, color=colors[i], label=f"rho = {r}")
plt.ylabel('Total Energy')
plt.xlabel('Temperature')
plt.legend()
plt.xlim(2, 6)
plt.ylim(-2, 12)
```
